```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
test=pd.read_csv('TEST.csv')
dataset=pd.read_csv('TRAIN.csv')
```

```python
from datetime import datetime
from pytz import timezone
import pytz
hawaii= timezone('Pacific/Honolulu')
dataset.index =  pd.to_datetime(dataset['UNIXTime'], unit='s')
dataset.index = dataset.index.tz_localize(pytz.utc).tz_convert(hawaii)
dataset['MonthOfYear'] = dataset.index.strftime('%m').astype(int)
dataset['DayOfYear'] = dataset.index.strftime('%j').astype(int)
dataset['WeekOfYear'] = dataset.index.strftime('%U').astype(int)
dataset['TimeOfDay(h)'] = dataset.index.hour
dataset['TimeOfDay(m)'] = dataset.index.hour*60 + dataset.index.minute
dataset['TimeOfDay(s)'] = dataset.index.hour*60*60 + dataset.index.minute*60 + dataset.index.second
dataset['TimeSunRise'] = pd.to_datetime(dataset['TimeSunRise'], format='%H:%M:%S')
dataset['TimeSunSet'] = pd.to_datetime(dataset['TimeSunSet'], format='%H:%M:%S')
dataset['DayLength(s)'] = dataset['TimeSunSet'].dt.hour*60*60 \
                           + dataset['TimeSunSet'].dt.minute*60 \
                           + dataset['TimeSunSet'].dt.second \
                           - dataset['TimeSunRise'].dt.hour*60*60 \
                           - dataset['TimeSunRise'].dt.minute*60 \
                           - dataset['TimeSunRise'].dt.second
dataset.drop(['Data','Time','TimeSunRise','TimeSunSet'], inplace=True, axis=1)
dataset.head()
```

```python
from datetime import datetime
from pytz import timezone
import pytz
hawaii= timezone('Pacific/Honolulu')
test.index =  pd.to_datetime(test['UNIXTime'], unit='s')
test.index = test.index.tz_localize(pytz.utc).tz_convert(hawaii)
test['MonthOfYear'] = test.index.strftime('%m').astype(int)
test['DayOfYear'] = test.index.strftime('%j').astype(int)
test['WeekOfYear'] = test.index.strftime('%U').astype(int)
test['TimeOfDay(h)'] = test.index.hour
test['TimeOfDay(m)'] = test.index.hour*60 + test.index.minute
test['TimeOfDay(s)'] = test.index.hour*60*60 + test.index.minute*60 + test.index.second
test['TimeSunRise'] = pd.to_datetime(test['TimeSunRise'], format='%H:%M:%S')
test['TimeSunSet'] = pd.to_datetime(test['TimeSunSet'], format='%H:%M:%S')
test['DayLength(s)'] = test['TimeSunSet'].dt.hour*60*60 \
                           + test['TimeSunSet'].dt.minute*60 \
                           + test['TimeSunSet'].dt.second \
                           - test['TimeSunRise'].dt.hour*60*60 \
                           - test['TimeSunRise'].dt.minute*60 \
                           - test['TimeSunRise'].dt.second
test.drop(['Data','Time','TimeSunRise','TimeSunSet'], inplace=True, axis=1)
test.head()
```

```python
X = dataset[['Temperature', 'Pressure', 'Humidity', 'WindDirection(Degrees)', 'Speed', 'DayOfYear', 'TimeOfDay(s)']]
y = dataset['Radiation']
```

```python
test_x = test[['Temperature', 'Pressure', 'Humidity', 'WindDirection(Degrees)', 'Speed', 'DayOfYear', 'TimeOfDay(s)']]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
regressor = RandomForestRegressor(n_estimators = 100)
```

```python
X_train_best = X[['Temperature','DayOfYear', 'TimeOfDay(s)']]
X_test_best = test_x[['Temperature', 'DayOfYear', 'TimeOfDay(s)']]
regressor.fit(X_train_best, y)
```

```python
from sklearn.metrics import explained_variance_score, mean_squared_error, r2_score
y_pred = regressor.predict(X_test_best)
```
